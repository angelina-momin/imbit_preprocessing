```python
pattern = re.compile(r"^[SIR]\d{5}$")
matched_pop_group_cols = [col for col in gdf_startpop.columns if pattern.match(col)]
list_cols_display = ["Name", "TotPop"] + matched_pop_group_cols

gdf_startpop = gdf_startpop[list_cols_display]

gdf_startpop["SumPop"] = gdf_startpop[matched_pop_group_cols].sum(axis=1)  

gdf_startpop["Check"] = gdf_startpop["TotPop"] == gdf_startpop["SumPop"]
gdf_startpop["diffSumPop"] = gdf_startpop["TotPop"] - gdf_startpop["SumPop"]
```
