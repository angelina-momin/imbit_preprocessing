```python
# Function to load and filter the CSV file
def filter_animal_infections(csv_file):
    # Load the CSV file with ';' as the delimiter
    df = pd.read_csv(csv_file, skiprows=[0,1], delimiter=';')

    # Strip column names of whitespace (useful for cleaner handling)
    df.columns = df.columns.str.strip()
    
    # Filter rows to show only cases we have animal infections
    filtered_df = df[
        (df['totalnewinfected'] > 0) 
    ]

    return filtered_df
```
